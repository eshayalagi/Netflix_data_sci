```python
import pandas as pd

# Step 1: Count of missing values per column
null = df.isna().sum().sort_values(ascending=False)

# Step 2: Percentage of missing values per column
percent = round(df.isna().mean() * 100, 2).sort_values(ascending=False)

# Step 3: Combine both into one table
missing_data = pd.concat([null, percent], axis=1, keys=['Missing Value Count', 'Percentage'])

# Step 4: Show only columns with missing values
missing_data = missing_data[missing_data['Missing Value Count'] != 0]

# Step 5: Print it
print("📊 Missing Value Summary Table:")
print(missing_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example: If you're loading data from a CSV
# df = pd.read_csv("your_file.csv")

# Step 1: Calculate missing value percentages
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_percent = missing_percent[missing_percent > 0].sort_values(ascending=False)

# Step 2: Plot as bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=missing_percent.index, y=missing_percent.values, palette="pastel")

# Step 3: Add labels and styling
plt.title("📊 Percentage of Missing Values per Column", fontsize=14)
plt.ylabel("Missing Value Percentage (%)", fontsize=12)
plt.xlabel("Column Name", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 100)  # Since max % is 100
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of bars
for i, value in enumerate(missing_percent.values):
    plt.text(i, value + 1, f"{value:.1f}%", ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

# FILLING THE MISSING VALUES

### 1️⃣ director: 2634 missing
#### This means: 2634 shows/movies have no director mentioned

#### Type: Categorical (text)

#### We can’t guess the correct director

#### So we fill it with something safe, like 'Unknown'

```python
df['director'].fillna('Unknown', inplace=True)
```

## 2️⃣ cast: 825 missing
#### Means: 825 shows have no cast listed
#### Type: Categorical (text)
#### It’s OK to not have cast sometimes
#### Fill with 'Unavailable' or 'Not Mentioned'



```python
df['cast'].fillna('Unavailable', inplace=True)
```

### 3️⃣ country: 831 missing
#### Means: We don’t know where 831 shows were produced
#### Type: Categorical
#### You can fill it with:'Unknown'
#### OR most common country (like 'United States')

```python
df['country'].fillna('Unknown', inplace=True)
```

### 4️⃣ date_added: 10 missing
#### Means: We don’t know when 10 shows were added to Netflix
#### Type: Date (like March 5, 2018)
#### We can use the most common date (called mode)

```python
df['date_added'].fillna(df['date_added'].mode()[0], inplace=True)
```

### 5️⃣ rating: 4 missing
#### Means: No age rating for 4 shows (like PG, TV-MA)
#### Type: Categorical
#### Best fix: Fill with the most common rating


```python
df['rating'].fillna(df['rating'].mode()[0], inplace=True)
```

### 6️⃣ duration: 3 missing
#### Means: 3 rows don’t have a duration (like "90 min" or "1 Season")
#### Type: Categorical (because it's text)
#### Solution: Fill with the most common duration

```python
df['duration'].fillna(df['duration'].mode()[0], inplace=True)
```
